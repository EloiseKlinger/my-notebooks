```python
import numpy as np
import matplotlib.pyplot as plt

def fonction_inverse(x):
    return 1 / x

# Création des points pour le graphique
x = np.linspace(-10, 10, 1000)
x = x[x != 0]  # Suppression du point x=0 car la fonction n'est pas définie en 0
y = fonction_inverse(x)

# Création du graphique
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='f(x) = 1/x')
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.title('Graphique de la fonction inverse')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Ajustement des limites pour mieux voir l'hyperbole
plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Définir les valeurs de x
x = np.linspace(-10, 10, 500)  # Génère 500 points entre -10 et 10

# Calculer y = x^3
y = x**3

# Tracer le graphique
plt.plot(x, y, label="y = x³", color="purple")

# Ajouter des titres et des légendes
plt.title("Graphique de la fonction y = x³")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # Ligne horizontale à y=0
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Ligne verticale à x=0
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()

# Afficher le graphique
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def racine_carree(x):
    return np.sqrt(x)

# Création des points pour le graphique
x = np.linspace(0, 10, 1000)  # On commence à 0 car la racine carrée n'est pas définie pour les nombres négatifs
y = racine_carree(x)

# Création du graphique
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r-', label='f(x) = √x')
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.title('Graphique de la fonction racine carrée')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Ajustement des limites pour mieux voir la courbe
plt.xlim(0, 10)
plt.ylim(0, 4)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Définir les valeurs de x
x = np.linspace(-2, 2, 50)  # Génère 50 points entre -2 et 2

# Calculer les fonctions
y1 = x**3
y2 = x**2
y3 = x

# Tracer les graphiques
plt.plot(x, y1, label="y = x³", color="blue")
plt.plot(x, y2, label="y = x²", color="green")
plt.plot(x, y3, label="y = x", color="red")

# Ajouter des titres et des légendes
plt.title("Graphique des fonctions y = x³, y = x² et y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # Ligne horizontale à y=0
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Ligne verticale à x=0
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()

# Afficher le graphique
plt.show()
```
