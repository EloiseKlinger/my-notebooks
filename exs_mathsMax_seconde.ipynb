Exercice 22 chapitre 9, Math Max

```python
# ex 22_chap IX (Math Max)
import matplotlib.pyplot as plt

def f(x):
    if -2 <= x <= 1: #1er cas
        return x - 1
    elif 1 < x <= 4: #2ème cas
        return -x**2 + 1
    else:
        return None  # Valeur n'est pas dans l'intervalle défini


    # Préparation des données pour le tracé
x_values = []
y_values = []
x = -2
while x <= 4:
    y = f(x)
    if y is not None:  # On ne trace pas les valeurs "None"
        x_values.append(x)
        y_values.append(y)
    x += 0.5  # Pas plus petit pour une courbe plus lisse

# Tracé de la courbe
plt.plot(x_values, y_values)

# Ajout des axes
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Titre et labels
plt.title("Courbe représentative de f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")

# Grille
plt.grid(True)

# Affichage du graphique
plt.show()
```

Exercice 4 chapitre 10, Math Max

```python
# ex 3_chap X (Math Max)
import numpy as np
import matplotlib.pyplot as plt

# Définition des valeurs de x
x = np.linspace(-3, 3, 300)  # Intervalle de x de -3 à 3 avec 300 points

# Définition des fonctions
f_x = x**2 - 4
g_x = -x - 2

# Tracé des courbes
plt.plot(x, f_x, label=r"$f(x) = x^2 - 4$", color="orange")
plt.plot(x, g_x, label=r"$g(x) = -x - 2$", color="red")

# Ajout des axes
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Ajout des éléments graphiques
plt.xlabel("x")
plt.ylabel("y")
plt.title("Représentation des fonctions $f(x) = x^2 - 4$ et $g(x) = -x - 2$")
plt.legend()
plt.grid()

# Affichage du graphique
plt.show()
```
