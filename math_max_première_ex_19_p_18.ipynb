**Exercice 19 p.18, Math Max première (modifié)**
Soit $b$ un réel et soit $P$ défini sur $\mathbb{R}$ par  
$$
P(x) = -\frac{1}{2}x^2 + bx - 3.
$$


- Créer un curseur $b$ allant de $-8$ à $8$ puis tracer la fonction $P$,et matérialiser le sommet.
- Faire varier le curseur $b$ puis démontrer votre conjecture, et proposer une conjecture sur le lieu géométrique des sommets.
- Démontrer cette conjecture.


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
```

```python
def P(x, b):
    return -0.5 * x**2 + b * x - 3

def sommet_parabole(b):
    # Pour P(x) = a x^2 + b x + c, sommet en x_s = -b/(2a)
    a = -0.5
    x_s = -b / (2 * a)
    y_s = P(x_s, b)
    return x_s, y_s
```

```python
def trace_parabole(b):
    x = np.linspace(-15, 15, 400)
    y = P(x, b)
    x_s, y_s = sommet_parabole(b)

    plt.figure(figsize=(8,5))
    plt.plot(x, y, label=f"$P(x) = -\\frac{{1}}{{2}}x^2 + {b:.2f}x - 3$")
    plt.scatter([x_s], [y_s], color='red', zorder=5, label=f"Sommet ({x_s:.2f}, {y_s:.2f})")
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.title("Parabole et déplacement du sommet selon b")
    plt.xlabel("x")
    plt.ylabel("P(x)")
    plt.legend()
    plt.grid(True)
    plt.ylim(-20, 20)
    plt.show()
```

```python
interact(trace_parabole, b=FloatSlider(min=-8, max=8, step=0.1, value=0))
```

```python
b_vals = np.linspace(-8, 8, 200)
x_s = b_vals
y_s = 0.5 * b_vals**2 - 3

plt.figure(figsize=(8,5))
plt.plot(x_s, y_s, label="Trajectoire du sommet")
plt.title("Lieux géométrique du sommet quand b varie")
plt.xlabel("x_s (abscisse du sommet)")
plt.ylabel("y_s (ordonnée du sommet)")
plt.grid(True)
plt.legend()
plt.show()
```




**Démonstration de la conjecture :**

On considère la fonction $P(x) = -\frac{1}{2}x^2 + bx - 3$ où $b$ est un réel.

Le sommet $S$ d'une parabole d'équation $ax^2 + bx + c$ a pour abscisse :
$$
x_S = -\frac{b}{2a}
$$

Ici, $a = -\frac{1}{2}$ et $b$ est le paramètre variable.  
Donc :
$$
x_S = -\frac{b}{2 \times \left(-\frac{1}{2}\right)} = -\frac{b}{-1} = b
$$

L'ordonnée du sommet est :
$$
y_S = P(x_S) = -\frac{1}{2}x_S^2 + b x_S - 3
$$
En remplaçant $x_S$ par $b$ :
$$
y_S = -\frac{1}{2}b^2 + b \times b - 3 = -\frac{1}{2}b^2 + b^2 - 3 = \frac{1}{2}b^2 - 3
$$

**Conclusion :**

Lorsque $b$ varie, le sommet $S$ de la parabole décrit la courbe d'équation :
$$
(x_S, y_S) = \left(b, \frac{1}{2}b^2 - 3\right)
$$

Autrement dit, le lieu des sommets est une parabole d'équation $y = \frac{1}{2}x^2 - 3$.
