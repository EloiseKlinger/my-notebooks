## L'impact limité des mutations

*L'impact limité des mutations*, activité pratique n°3, page 87 de :

SVT (Spécialité Terminale) 2020 Manuel de lélève (collection Baude & Jusserand, Bordas)



```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres de simulation
population_size = 1000  # Taille de la population
num_generations = 100    # Nombre de générations
mutation_rate = 0.01     # Taux de mutation (1%)

# Initialisation des fréquences des allèles
allele_A = 0.5  # Fréquence initiale de l'allèle A
allele_B = 1 - allele_A  # Fréquence initiale de l'allèle B

# Listes pour stocker les fréquences au fil des générations
frequencies_A = []
frequencies_B = []

for generation in range(num_generations):
    # Calculer le nombre d'individus avec chaque allèle
    num_A = int(population_size * allele_A)
    num_B = int(population_size * allele_B)

    # print("num_A : ", num_A)
    # print("num_B : ", num_B)

    # Appliquer les mutations
    mutations_to_B = np.random.binomial(num_A, mutation_rate)
    mutations_to_A = np.random.binomial(num_B, mutation_rate)

    # Mettre à jour les fréquences des allèles après mutations
    num_A -= mutations_to_B
    num_B += mutations_to_B
    num_B -= mutations_to_A
    num_A += mutations_to_A

    # Recalculer les fréquences
    total_population = num_A + num_B
    allele_A = num_A / total_population if total_population > 0 else 0
    allele_B = num_B / total_population if total_population > 0 else 0

    # Stocker les fréquences pour le traçage
    frequencies_A.append(allele_A)
    frequencies_B.append(allele_B)

# Traçage des résultats
plt.plot(frequencies_A, label='Fréquence Allèle A')
plt.plot(frequencies_B, label='Fréquence Allèle B')
plt.title('Variation de la Fréquence des Allèles au Cours du Temps')
plt.xlabel('Générations')
plt.ylabel('Fréquence')
plt.legend()
plt.grid()
plt.show()
```
